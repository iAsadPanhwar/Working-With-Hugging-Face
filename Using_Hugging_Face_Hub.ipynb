```python
from huggingface_hub import HfApi, ModelFilter
```

```python
# @title
api = HfApi()
list(api.list_models())
```

**Searching A Model For a task**

```python
models = api.list_models(
    filter = ModelFilter(
        task = "text-classification"),

        sort = "downloads",
        direction = -1,
        limit = 5


)

modellist = list(models)

print(modellist[0])
```

```python
models = api.list_models(
    filter = ModelFilter(
        task = "text-generation"
    ),
    sort = 'downloads',
    direction = -1,
    limit = 5
)

modellist = list(models)

print(modellist[0:2])
```

**Saving a model Locally**

```python
from transformers import AutoModel

modelId = "distilbert-base-uncased-finetuned-sst-2-english"

#Download model using the modelId
model = AutoModel.from_pretrained(modelId)

#Save the model to a local directory
model.save_pretrained(save_directory = f'models/{modelId}')
```

**Working With Datasets**

```python
from datasets import load_dataset
```

```python
data = load_dataset('imdb', split = 'train')
```

```python
filtered = data.filter(lambda row: row['label'] == 0)
```

```python
filtered
```

```python
sliced = filtered.select(range(2))
```

```python
print(sliced)
```

```python
print(sliced[0]['text'])
```

```python
wikipedia = load_dataset('wikimedia/wikipedia', split = 'train',  name = "20231101.en")

print(f'Length of the dataset: {len(wikipedia)}')
```

```python
filtered = wikipedia.filter(lambda row: 'football' in row['text'])

example = filtered.select(range(1))

print(example[0]['text'])
```
